# AN√ÅLISIS OBRAS

üìäüìà **Para realizar un peque√±o an√°lisis de las obras se realizar√°n un par de gr√°ficos.**

1Ô∏è‚É£ En el primero, se muestra cual es aproximadamente la media de obras escritas por un autor, es decir, cuantas obras es m√°s com√∫n que realice un mismo autor. Adem√°s, esta informaci√≥n vendr√° acompa√±ada de un describe donde se pueden observar el n√∫mero medio de obras, el n√∫mero m√°ximo o el m√≠nimo entre otos datos de inter√©s. 

2Ô∏è‚É£ Por otro lado, un gr√°fico que aporta informaci√≥n sobre el n√∫mero de obras perteneciente a cada g√©nero. De este modo, es posible analizar los g√©neros que poseen un mayor n√∫mero de obras.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
art = pd.read_csv('../data/art.csv', index_col = 0)
art.head(2)
```

```python
art.describe()
```

```python
obras = sns.displot(art.paintings,
            color='navy', 
            edgecolor = "black", 
            bins= 50, 
            kde=True, 
            height = 10,
            aspect =0.8)

obras.set(title = "Total n√∫mero de obras m√°s com√∫n")
obras.savefig('../images/Total n√∫mero de obras m√°s com√∫n.png');
```

ü§îü§ì De este modo, podemos apreciar que la mayor√≠a de autores escriben menos de 200 obras. Junto a la informaci√≥n aportada por el describe, podemos afirmar que 191 obras se posiciona en el tercer cuartil, lo que nos indica que un 75% de los autores han realizado 191 obras o menos. 

```python
plt.rcParams["figure.figsize"] = (20, 8)
genres = art['genre']


fig2 = sns.lineplot(data= art, x="genre", y="paintings",
                    color = "black",
                    marker = 'o',
                    markersize = 8,
                    linewidth = 2);

fig2.set_facecolor("white")
fig2.spines['bottom'].set_color('black') 
fig2.spines['top'].set_color('white') 
fig2.spines['right'].set_color('white')
fig2.spines['left'].set_color('black')

plt.ylim(0,500)

fig2.set_xticklabels(genres, rotation = 55)

plt.title("N√∫mero de obras por g√©nero")
                     
plt.savefig("../images/N√∫mero de obras por g√©nero.png")
    
plt.show();
```

ü§®üßê En este caso, podemos observar como el cubirmo, en primer lugar, y el realismo social, en segundo, destacan como los g√©neros con el mayor n√∫mero de obras. Por su parte, el impresionismo destaca por el poco n√∫mero de obras.
